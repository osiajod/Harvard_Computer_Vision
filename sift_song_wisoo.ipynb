## 1. Detecting SIFT Interest Points

OpenCV includes good implementations of a variety interest point detectors and descriptors. Its documentation on this topic is also relatively good, and the set of tutorials is recommended reading, both for learning and as a reference for use:

[OpenCV-Python Tutorials: Feature Detection and Description](https://docs.opencv.org/master/db/d27/tutorial_py_table_of_contents_feature2d.html)

The notebook you are working on now deals with the detector and descriptor known as SIFT (Scale-Invariant Feature Transform).

### Instructions

2. Read the [SIFT portion of the OpenCV tutorial](https://docs.opencv.org/master/da/df5/tutorial_py_sift_intro.html), paying particular attention to the Section titled **SIFT in OpenCV**.
3. Following the instructions in that tutorial, edit lines 20 and 23 below to pass a flag to the `drawKeypoints()` function that changes the visualization to include the size (scale) and orientation of each detected interest point.

```python
# Load two images and convert to grayscale
im1 = cv2.cvtColor(cv2.imread('./quad/quad_left.jpg'), cv2.COLOR_BGR2RGB)
im2 = cv2.cvtColor(cv2.imread('./quad/quad_middle.jpg'), cv2.COLOR_BGR2RGB)
im1_gray = cv2.cvtColor(im1, cv2.COLOR_RGB2GRAY)
im2_gray = cv2.cvtColor(im2, cv2.COLOR_RGB2GRAY)

# Create a SIFT object
sift = cv2.SIFT_create()

# Use the detectAndCompute method of this SIFT object to 
#  detect SIFT keypoints and compute descriptor for each one
keypoints1, descriptors1 = sift.detectAndCompute(im1_gray, None)
keypoints2, descriptors2 = sift.detectAndCompute(im2_gray, None)

# Create new images that include visualizations of the detected interest points
# Note: Due to a quirk in openCV, the third argument (which the documentation
#       says should be the output image) is ignored so can be set to None.
imKeypoints1=cv2.drawKeypoints(im1,
                              keypoints1,
                              None,
                              flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
                              )
imKeypoints2=cv2.drawKeypoints(im2,
                              keypoints2,
                              None,
                              flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
                              )

f, (ax1,ax2) = plt.subplots(2,1,figsize=(15,20))
ax1.imshow(imKeypoints1)
ax1.axis('off')
ax2.imshow(imKeypoints2)
ax2.axis('off')
```
