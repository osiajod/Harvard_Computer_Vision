```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')   # Load image
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)                 # Plot image
plt.axis('off')                 # Remove axes from figure
plt.title('My image');          # Create title
```

```python
print(f'img_rgb has data type {img_rgb.dtype}.')
```

```python
print(f'The minimal value in img_rgb is {img_rgb.min()} and the maximal value is {img_rgb.max()}')
```

```python
img_float32 = np.float32(img_rgb) / 255.0
```

```python
print(f'img_float32 has data type {img_float32.dtype}.')
```

```python
img_gray = cv2.cvtColor(img_float32, cv2.COLOR_RGB2GRAY)
```

```python
print(f'Original image shape: {img_float32.shape}. Grayscale image shape: {img_gray.shape}.')
```

```python
plt.imshow(img_gray, cmap='gray')   # Plot image
plt.axis('off')                     # Remove axes from figure
plt.title('My image - grayscale');  # Create title
```

```python
cropped_grayscale_img = img_gray[20:-10, 60:-16]
plt.imshow(1- cropped_grayscale_img, cmap='gray')
plt.axis('off')
plt.title("My Image - cropped and inverted")
```

```python
rotated = cv2.rotate(cropped_grayscale_img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(rotated, cmap='gray')
plt.axis('off')
plt.title("90 degrees rotation, clockwise")
```
